#繪製函數圖形

$$ x = 16 * np.sin(t)**3 \\y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t) $$

```python
t = np.linspace(0, 2 * np.pi, 400)  # 初始數組 t 並填充均勻點在 [0, 2pi] 區間
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# 使用模組:Pyplot 繪製愛心
plt.figure(figsize=(6, 6))  # 設置圖形大小(單位英尺)  figure(figsize=(WIDTH_SIZE寬,HEIGHT_SIZE高))
plt.plot(x, y, color='blue', linewidth=10)  # 設置線條顏色
plt.fill(x, y, color='blue', alpha=0.3)  # 填充內部顏色
plt.title("loveloveloveStretch")  # 添加標題
```

# `plt.axis()`

- Matplotlib 自帶的函數，用於得到目前繪製圖形座標軸範圍
- 括號為空，即不帶任何參數時，返回的是目前座標軸範圍
---

# `axis_range` 是什麼？

`axis_range` 是一個 **元組（tuple）**。

當調用 `plt.axis()` 時，它會返回一個包含四個浮點數的元組，格式為 `(xmin, xmax, ymin, ymax)`，分別表示當前坐標軸的範圍：

- `xmin`：x 軸的最小值
- `xmax`：x 軸的最大值
- `ymin`：y 軸的最小值
- `ymax`：y 軸的最大值

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2 * np.pi, 400)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

plt.figure(figsize=(6, 6))
plt.plot(x, y, color='blue', linewidth=10)
plt.fill(x, y, color='blue', alpha=0.3)
plt.title("loveloveloveStretch")
plt.axis()
axis_range = plt.axis() # 將 plt.axis() 返回的內容用 axis_range 存儲並輸出
print("當前坐標軸範圍 (xmin, xmax, ymin, ymax):", axis_range)
```

可以注意到，範圍並非我所想的都是一樣的大小，當前坐標軸範圍

- (-17.599590839252688, 17.599590839252688,
- -18.445123375510324, 13.368730389996315)

#手動設置坐標軸範圍

# `max(abs(x_min), abs(x_max), abs(y_min), abs(y_max)`)
：從這些絕對值中找出最大的值。
- 通過計算出一個統一的範圍（axis_range），並將其應用於 x 軸和 y 軸，可以確保圖形在水平和垂直方向上等比例顯示
- 圖形會顯示為四方等長

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2 * np.pi, 400)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

plt.figure(figsize=(6, 6))
plt.plot(x, y, color='red', linewidth=10)
plt.fill(x, y, color='red', alpha=0.3)
plt.title("lovelovelove")

# 計算坐標軸範圍
x_min, x_max = min(x), max(x)
y_min, y_max = min(y), max(y)

# 使用 plt.axis() 設置四方等長的範圍
axis_range = max(abs(x_min), abs(x_max), abs(y_min), abs(y_max))
plt.axis([-axis_range, axis_range, -axis_range, axis_range])  # 設置 x 和 y 軸範圍

print("當前坐標軸範圍 (xmin, xmax, ymin, ymax):", plt.axis())
```

#比較和總結
